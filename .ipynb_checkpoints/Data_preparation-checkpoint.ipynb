```python
# Loading Data
import pandas as pd
import numpy as np
train_file = pd.read_csv("dataset/train.csv")
test_file = pd.read_csv("dataset/test.csv")
stores_file = pd.read_csv("dataset/stores.csv")
sample_file = pd.read_csv("dataset/sampleSubmission.csv")
features_file = pd.read_csv("dataset/features.csv")
```

```python
from IPython.core.display import HTML
#display(HTML(stores_file.to_html()))
```

```python
subset_A = stores_file[stores_file['Type'] == 'A']
subset_B = stores_file[stores_file['Type'] == 'B']
subset_C = stores_file[stores_file['Type'] == 'C']
#display(HTML(subset_A.to_html()))
```

```python
df_storeA = subset_A.sample(n=3)
df_storeB = subset_B.sample(n=3)
df_storeC = subset_C.sample(n=3)
frames = [df_storeA, df_storeB, df_storeC]
result = pd.concat(frames)
#display(HTML(result.to_html()))
```
